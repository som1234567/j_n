```python
import itertools
letters = ['a', 'b', 'c', 'd']
numbers = [0, 1, 2, 3, 4]
names = ['sudhan', 'Rhythm']

result1 = itertools.permutations(letters, 2) # repeats
result2 = itertools.combinations_with_replacement(numbers, 4)
result3 = itertools.combinations(letters, 2) # gives different ... do not repeat


for item in result1:
    print(item)
```

```python
result = itertools.product(numbers, repeat=4)
for item in result:
    print(item)
```

# loop over all the different items.... in chain

```python
combined = itertools.chain(letters, numbers, names)
for item in combined:
    print(item)
```

# islice:: 

```python
result = itertools.islice(range(10), 5) # stops at 5
result1 = itertools.islice(range(10),1, 5) # # begins from 1 and stops at 
result2 = itertools.islice(range(10),1, 5, 2)# begins from 1 and stops at 5 steps with 2
for item in result1:
    print(item)
```

```python
import itertools
with open('test.txt', 'r') as f:
    header = itertools.islice(f, 3) # takes just 3 lines of the file to save the memory
    
    for line in header:
        print(line, end='')
```

```python
letters = ['a', 'b', 'c', 'd']
selectors = [True, False, False, True]
result = itertools.compress(letters, selectors) # used to filter the values
for item in result:
    print(item)
```
